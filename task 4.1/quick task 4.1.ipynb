```python
import csv

with open('covid_19_data.csv', 'r') as f:
    c = csv.reader(f)
    data = [i for i in c]
data
dict_data = {}

# Країни та кількість людей з виявленим вірусом
for row in data[1:]:
    dict_data[row[3]] = row[5]
dict_data
    
```

```python
import pickle

with open(r'infected_people.pickle', 'wb') as f:
    pickle.dump(dict_data, f)
```
